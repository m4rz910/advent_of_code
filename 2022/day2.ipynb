```python
import pandas as pd
```

# part 1

```python
def points(s):
    hands = {"A":"Rock","B":"Paper","C":"Scissors", # opponent
             "X":"Rock","Y":"Paper","Z":"Scissors"} # me
    shape_value = {"Rock":1,"Paper":2,"Scissors":3} # my points
    outcome_value = {'Loss':0,"Tie":3,"Win":6}
    
    opponenthand = hands[s['opponenthand']]
    myhand = hands[s['myhand']]
    
    if opponenthand == myhand:
        outcome = 'Tie'
    else:
        if opponenthand == "Rock":
            if myhand == "Paper":
                outcome = "Win"
            else:
                outcome = "Loss"
        elif opponenthand == "Paper":
            if myhand == "Scissors":
                outcome = "Win"
            else:
                outcome = "Loss"
        elif opponenthand == "Scissors":
            if myhand == "Rock":
                outcome = "Win"
            else:
                outcome = "Loss"
        else:
            raise Exception(f"Error {opponenthand} does not exist")
    
    return outcome, outcome_value[outcome] + shape_value[myhand]  
    
df = pd.read_csv("data/day2.csv",header=None).rename(columns={0:'opponenthand',1:'myhand'})
    
df['mypoints'] = df.apply(lambda s: points(s),
                          axis='columns')
df['outcome'] = df['mypoints'].apply(lambda x: x[0])
df['points'] = df['mypoints'].apply(lambda x: x[1])
df
```

```python
df['points'].sum()
```

# part 2

```python
def points(s):
    hands = {"A":"Rock","B":"Paper","C":"Scissors"} # opponent
    outcomes = {"X":"Loss","Y":"Tie","Z":"Win"} # me
    
    shape_value = {"Rock":1,"Paper":2,"Scissors":3} # my points
    outcome_value = {'Loss':0,"Tie":3,"Win":6}
    
    opponenthand = hands[s['opponenthand']]
    #myhand = hands[s['myhand']] #needs to be determined
    
    outcome = outcomes[s['outcome']]
    if outcome == "Tie":
        myhand = opponenthand
    elif outcome == "Loss":
        if opponenthand == "Scissors":
            myhand = "Paper"
        elif opponenthand == "Rock":
            myhand = "Scissors"
        elif opponenthand == "Paper":
            myhand = "Rock"
        else:
            raise Exception("Error")
    elif outcome == "Win":
        if opponenthand == "Scissors":
            myhand = "Rock"
        elif opponenthand == "Rock":
            myhand = "Paper"
        elif opponenthand == "Paper":
            myhand = "Scissors"
        else:
            raise Exception("Error")
    else:
        raise Exception(f"Error {outcome} does not exist")
    
    return myhand, outcome_value[outcome] + shape_value[myhand]  
    
df = pd.read_csv("data/day2.csv",header=None).rename(columns={0:'opponenthand',1:'outcome'})
    
df['mypoints'] = df.apply(lambda s: points(s),
                          axis='columns')
df['outcome'] = df['mypoints'].apply(lambda x: x[0])
df['points'] = df['mypoints'].apply(lambda x: x[1])
df
```
