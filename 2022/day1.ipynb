```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/day1.csv', skip_blank_lines=False, header=None, low_memory=False)
calories = df.values.flatten()

data = {}

elf = 0
for i,c in enumerate(calories):
    try:
        data[elf]
    except KeyError: # create the list if it doesnt exist
        data[elf] = []
        
    if np.isnan(c):
        elf = elf + 1
        continue
    else:
        data[elf].append(c)

print(f"Number of Elfs: {elf}")      
df = pd.DataFrame.from_dict(data,orient='index').T
df
```

```python
print(f"Elf: {df.sum().idxmax()} has {df.sum().max()} calories")
```
